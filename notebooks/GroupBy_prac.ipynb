```python
import pandas as pd
df = pd.read_csv("incubator.csv")
df
```

```python
# group data with "BirdType"
g = df.groupby("Bird")
g
```

```python
for bird_type, bird_type_df in g:
    print(bird_type)
    print(bird_type_df)
    print()
```

```python
# get 'turkey' group data
# ~returns a df
g.get_group("turkey")
```

```python
# get max values on the grouped df
print("[INFO] Maximum values")
g.max()
```

```python
# get average for each of the bird type
print("Average Values")
g.mean()
```

```python
# get all stats with g.describe
g.describe()
```
